Neste exercício vamos trabalhar com a base "Heart Disease Data Set" disponível no site do [UCI](https://archive.ics.uci.edu/ml/datasets/heart+disease).

Temos 303 indivíduos e vamos utilizar 13 variáveis disponíveis para prever a presença de doença cardíaca na variável ```num```(a 14$^a$ variável).

A descrição das variáveis está recortada abaixo:
- age: idade do paciente em anos
- sex: sexo (1 = male; 0 = female)  
- cp: tipo de dor no peito
  - 1: angina típica
  - 2: angina atípica
  - 3: dor não-angina
  - 4: assintomático
- trestbps: pressão sanguínea em repouso (em mm Hg na admissão ao hospital
- chol: colesterol sérico em mg/dl
- fbs: (açúcar no sangue em jejum > 120 mg/dl) (1 = True; 0 = False)
- restecg: resultados eletrocardiográficos em repouso
  - 0: normal
  - 1: tendo anormalidade da onda ST-T (Inversões de onda T e / ou ST com elevação ou depressão de > 0.05 mV)
  - 2: mostrando hipertrofia ventricular esquerda provável ou definitiva pelos critérios de Estes
- thalach: frequência cardíaca máxima alcançada
- exang: angina induzida por exercício(1 = sim; 0 = não)
- oldpeak = Depressão de ST induzida por exercício em relação ao repouso
- slope: Depressão de ST induzida por exercício em relação ao repouso
  - 1: inclinação ascendente
  - 2: estável
  - 3: inclinação descendente
- ca: número de vasos principais (0-3) coloridos por fluorosopia
- thal: 3 = normal; 6 = defeito corrigido; 7 = defeito reversível
- num: diagnóstico de doença cardíaga (status de doença angiográfica)

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.formula.api as smf
````

````python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'

df = pd.read_csv(url, 
                 names=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                        'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'])
df.head()
````

````python
df['flag_doente'] = (df['num'] != 0).astype('int64')
df
````

1. Monte uma bivariada para a variável ```sex``` com as seguintes estatísticas:

- Número de *saudáveis*
- Número de *doentes*
- Total
- Quantidade média de doentes
- *Odds*
- *Odds ratio* (Use o total como referência)
- Logito (LOG(*Odds*))
- O LOG do *Odds ratio* (com o total como referência) é chamado de *Weight of Evidence* (WOE). Calcule-o.

````python
# Calcular as estatísticas solicitadas
df['doente'] = df['flag_doente']
df['saudavel'] = np.where(df['flag_doente']==1, 0, 1)

# Calcule o número de saudáveis e doentes para cada sexo
resumo = df.groupby('sex')[['saudavel', 'doente']].sum()

# Calcule o total
resumo['total'] = resumo['saudavel'] + resumo['doente']

# Calcule a quantidade média de doentes
resumo['media_doentes'] = resumo['doente'] / resumo['total']

# Calcule as odds
resumo['odds'] = resumo['doente'] / resumo['saudavel']

# Calcule o odds ratio usando o total como referência
total_odds = resumo.loc[:, 'odds'].sum()
resumo['odds_ratio'] = resumo['odds'] / total_odds

# Calcule o logito (log das odds)
resumo['logito'] = np.log(resumo['odds'])

# Calcule o Weight of Evidence (WOE)
resumo['woe'] = np.log(resumo['odds_ratio'])

resumo
````

2. Monte uma bivariada para a variável ```age``` com as seguintes estatísticas:
- Categorize a variável em 5 grupos
- Calcule as mesmas quantidades calculadas no execício anterior
- Calcule também o valor médio da variável ```age```

````python

# Categorize a variável 'age' em 5 grupos
df['age_group'] = pd.qcut(df['age'], 5)

# Calcule o número de saudáveis e doentes para cada grupo de idade
resumo2 = df.groupby('age_group')[['saudavel', 'doente']].sum()

# Calcule o total
resumo2['total'] = resumo2['saudavel'] + resumo2['doente']

# Calcule a quantidade média de doentes
resumo2['media_doentes'] = resumo2['doente'] / resumo2['total']

# Calcule as odds
resumo2['odds'] = resumo2['doente'] / resumo2['saudavel']

# Calcule o odds ratio usando o total como referência
total_odds2 = resumo2.loc[:, 'odds'].sum()
resumo2['odds_ratio'] = resumo2['odds'] / total_odds2

# Calcule o logito (log das odds)
resumo2['logito'] = np.log(resumo2['odds'])

# Calcule o Weight of Evidence (WOE)
resumo2['woe'] = np.log(resumo2['odds_ratio'])

# Calcule o valor médio da variável 'age'
resumo2['media_age'] = df.groupby('age_group')['age'].mean()

resumo2
````

3. Visualize a informação da bivariada construída acima através de um gráfico da sua escolha.

````python
# Crie uma figura com 4 subplots
fig, axs = plt.subplots(4)

# Crie um gráfico de barras para a quantidade média de doentes
axs[0].bar(resumo2.index.astype(str), resumo2['media_doentes'])
axs[0].set_title('Quantidade média de doentes')

# Crie um gráfico de barras para as odds
axs[1].bar(resumo2.index.astype(str), resumo2['odds'])
axs[1].set_title('Odds')

# Crie um gráfico de barras para o odds ratio
axs[2].bar(resumo2.index.astype(str), resumo2['odds_ratio'])
axs[2].set_title('Odds Ratio')

# Crie um gráfico de barras para o WOE
axs[3].bar(resumo2.index.astype(str), resumo2['woe'])
axs[3].set_title('Weight of Evidence (WOE)')

# Mostre a figura
plt.tight_layout()
plt.show()
````
