```python
from pandas_datareader import data as pdr
import yfinance as yf

import plotly.express as px
from plotly import graph_objects

import pandas as pd
import numpy as np
```

```python
ativos =['PETR4.SA', 'ABEV3.SA', 'VALE3.SA']
dt_ini, dt_fim = '2023-01-01', '2023-11-30'

yf = pdr.get_data_yahoo(ativos, start=dt_ini, end=dt_fim)
yf.head()
```

```python
yf.to_csv('dados_yf.csv')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
yf.columns
```

```python
yf['amplitude_ABEV'] = yf[('High', 'ABEV3.SA')] - yf[('Low', 'ABEV3.SA')]
yf['amplitude_PETR4'] = yf[('High', 'PETR4.SA')] - yf[('Low', 'PETR4.SA')]
yf['amplitude_VALE3'] = yf[('High', 'VALE3.SA')] - yf[('Low', 'VALE3.SA')]

yf.head()
```

```python
# Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
fig = px.line(yf, x=yf.index, y=['amplitude_ABEV', 'amplitude_PETR4', 'amplitude_VALE3'], template='plotly_white')
fig.show()
```

Ativo apresentou maior volatilidade.
- VALE3
