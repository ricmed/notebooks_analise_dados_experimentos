# Probabilidade - Tarefa

```python
import numpy as np
from scipy.stats import norm, t
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Média (mu) e desvio padrão (sigma)
mu = 15
sigma = 0.1

# Valor de interesse
X = 15.2

# Calculando o score Z
Z = (X - mu) / sigma

# Calculando a probabilidade de estar acima de 15,2 kg
prob_acima_15_2 = 1 - norm.cdf(Z)

prob_acima_15_2
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Percentil desejado
percentil = 0.95

# Calculando o valor correspondente ao percentil 95
valor_percentil_95 = norm.ppf(percentil, mu, sigma)

valor_percentil_95
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
# Criando o vetor Y
Y = np.linspace(-6, 6, 200)

# Calculando as probabilidades para a distribuição Normal Padrão
prob_normal = norm.cdf(Y)

# Calculando as probabilidades para a distribuição t-Student com 5 graus de liberdade
prob_t_student = t.cdf(Y, df=5)

# Empacotando os resultados
resultados = {
    "Valores_Y": Y,
    "Prob_Normal_Padrão": prob_normal,
    "Prob_t-Student_5_graus": prob_t_student
}

resultados
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
# Criando o gráfico
plt.figure(figsize=(12, 8))

# Plotando as probabilidades para a distribuição Normal Padrão
plt.plot(Y, prob_normal, label="Normal Padrão", color="blue")

# Plotando as probabilidades para a distribuição t-Student com 5 graus de liberdade
plt.plot(Y, prob_t_student, label="t-Student (5 graus de liberdade)", color="red")

# Adicionando título e legendas
plt.title("Probabilidades Cumulativas para Distribuições Normal Padrão e t-Student")
plt.xlabel("Valores de Y")
plt.ylabel("Probabilidade Cumulativa")
plt.legend()

# Mostrando o gráfico
plt.grid(True)
plt.show()
```

5. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *Y* abaixo:

```python
# Criando o vetor Y para percentis
Y_percentis = np.linspace(0, 1, 20)

# Calculando os percentis para a distribuição Normal Padrão
percentis_normal = norm.ppf(Y_percentis)

# Calculando os percentis para a distribuição t-Student com 5 graus de liberdade
percentis_t_student = t.ppf(Y_percentis, df=5)

# Empacotando os resultados
resultados_percentis = {
    "Percentis": Y_percentis,
    "Percentis_Normal_Padrão": percentis_normal,
    "Percentis_t-Student_5_graus": percentis_t_student
}

resultados_percentis
```

6. Faça um gráfico com os dados do item anterior.

```python
# Criando o gráfico
plt.figure(figsize=(12, 8))

# Plotando os percentis para a distribuição Normal Padrão
plt.plot(Y_percentis, percentis_normal, label="Normal Padrão", color="blue")

# Plotando os percentis para a distribuição t-Student com 5 graus de liberdade
plt.plot(Y_percentis, percentis_t_student, label="t-Student (5 graus de liberdade)", color="red")

# Adicionando título e legendas
plt.title("Percentis para Distribuições Normal Padrão e t-Student")
plt.xlabel("Percentis")
plt.ylabel("Valor do Percentil")
plt.legend()

# Mostrando o gráfico
plt.grid(True)
plt.show()
```
